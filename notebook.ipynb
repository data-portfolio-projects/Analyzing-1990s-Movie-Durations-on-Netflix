## **Methods**

### **Data Preparation and Analysis**
- Import Libraries: Import pandas and matplotlib for data manipulation and visualization.
- Load Data: Read the Netflix CSV file into a DataFrame.
- Data Cleaning: Ensure the 'duration' column is in the correct format by removing ' min' and converting to integer.
- Filter Data: Filter the DataFrame for movies released in the 1990s.
- Calculate Mode: Determine the most frequent movie duration in the 1990s.
- Filter Specific Movies: Count the number of short action movies (duration < 90 minutes) released in the 1990s.

## **Code Implementation**

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Check the first few rows of the dataset to understand its structure
print(netflix_df.head())
```

```python
# Ensure the 'duration' column is in the correct format
# Remove ' min' from the 'duration' column and convert to integer

# Convert the 'duration' column to string first to avoid the AttributeError
netflix_df['duration'] = netflix_df['duration'].astype(str).str.replace(' min', '', regex=False).astype(int)
```

```python
# Filter for movies released in the 1990s
movies_1990s = netflix_df[(netflix_df['release_year'] >= 1990) & 
                          (netflix_df['release_year'] < 2000) & 
                          (netflix_df['type'] == 'Movie')]
```

```python
# What was the most frequent movie duration in the 1990s?
most_frequent_duration = movies_1990s['duration'].mode()[0]
duration = int(most_frequent_duration)
```

```python
# Count the number of short action movies released in the 1990s
short_action_movies = movies_1990s[(movies_1990s['duration'] < 90) & 
                                    (movies_1990s['genre'].str.contains('Action', case=False, na=False))]
```

```python
short_movie_count = short_action_movies.shape[0]
```

```python
# Print the results
print(f'The most frequent movie duration in the 1990s is: {duration} minutes')
print(f'The number of short action movies released in the 1990s is: {short_movie_count}')
```

```python
# Optional: Visualization of movie durations from the 1990s
plt.figure(figsize=(12, 6))
plt.hist(movies_1990s['duration'], bins=30, color='blue', alpha=0.7)
plt.axvline(x=duration, color='red', linestyle='dashed', linewidth=2, label=f'Most Frequent Duration: {duration} min')
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.legend()
plt.show()
```

## **Analysis of the Histogram**
